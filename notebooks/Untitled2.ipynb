```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from tqdm import tqdm_notebook as tqdm
```

```python
N = 500

T = int(1e3)

W = np.random.normal(0.,1.,(N,N))
W[range(N),range(N)] = 0.

x = np.ndarray((T,N))
x[0,:] = np.random.normal(0.,1.,(N))

r = 1.
```

```python
for t in tqdm(range(1,T)):
    x[t] = W @ x[t-1]
    x[t] = r * x[t]/np.linalg.norm(x[t])
```

```python
plt.figure()
plt.plot(x[:,:10])
plt.show()
```
